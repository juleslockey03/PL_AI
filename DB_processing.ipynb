# Imports and preprocessing

## Imports

```python
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder
```

## Teams DB

```python
data = pd.read_csv('data/england-premier-league-teams-2018-to-2019-stats.csv')
```

```python
data.head(5)
```

```python
start_data = data[['common_name','matches_played', 'wins', 'wins_home','wins_away','losses_home','losses_away', 'losses']]
```

```python
start_data['home_percentage'] = (start_data['wins_home']/19).round(2)
```

```python
start_data['away_percentage'] = (start_data['wins_away']/19).round(2)
```

```python
start_data['points'] = (data['points_per_game']*38).round(0)
```

```python
start_data.sort_values('points', ascending = False)
```

## Matches DB

```python
match_data = pd.read_csv('data/england-premier-league-matches-2018-to-2019-stats.csv')
```

```python
match_data.head(5)
```

```python
match_data_pre = match_data.drop('timestamp', axis = 1)
```

```python
match_data_pre = match_data.drop('home_ppg', axis = 1)
match_data_pre = match_data.drop('away_ppg', axis = 1)
```

```python
match_data_pre = match_data_pre[['home_team_name', 'away_team_name', 'home_team_goal_count', 'away_team_goal_count']]
```

```python
match_data_pre
```

```python
match_data_pre['away_team_away_percentage'] = 0
match_data_pre['home_team_home_percentage'] = 0
```

```python
for i in range(0, len(match_data_pre)):
    for j in range(0, len(start_data)):
        if match_data_pre['home_team_name'][i] == start_data['common_name'][j]:
            match_data_pre['home_team_home_percentage'][i] = start_data['home_percentage'][j]
            
        if match_data_pre['away_team_name'][i] == start_data['common_name'][j]:
            match_data_pre['away_team_away_percentage'][i] = start_data['away_percentage'][j]
```

```python
ohe = OneHotEncoder(sparse_output = False)
ohe.fit(match_data_pre[['home_team_name']])
```

```python
match_data_pre[ohe.get_feature_names_out()] = ohe.transform(match_data_pre[['home_team_name']])

match_data_pre.drop(columns = ["home_team_name"], inplace = True)
```

```python
ohe = OneHotEncoder(sparse_output = False)
ohe.fit(match_data_pre[['away_team_name']])
```

```python
match_data_pre[ohe.get_feature_names_out()] = ohe.transform(match_data_pre[['away_team_name']])

match_data_pre.drop(columns = ["away_team_name"], inplace = True)
```

```python
match_data_pre['home_win'] = 0
match_data_pre['away_win'] = 0
```

```python
for i in range(0,len(match_data_pre)):
    if match_data_pre['home_team_goal_count'][i] > match_data_pre['away_team_goal_count'][i]:
        match_data_pre['home_win'][i] = 1
        match_data_pre['away_win'][i] = 0
    elif match_data_pre['home_team_goal_count'][i] < match_data_pre['away_team_goal_count'][i]:
        match_data_pre['away_win'][i] = 1
        match_data_pre['home_win'][i] = 0

for i in range(0,len(match_data_pre)):
    if match_data_pre['home_win'][i] == 0 and match_data_pre['away_win'][i] == 0:
        match_data_pre['home_win'][i] = 0.5
        match_data_pre['away_win'][i] = 0.5
    
       
```

```python
match_data_pre['gameweek'] = match_data['Game Week']
```

# Beginning with the models

## Imports

```python
from tensorflow.keras import models, layers
```

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
```

## Early Tests

```python
mean_absolute_error_test_baseline = abs((y_train - y_train.mean())).mean()
mean_absolute_error_test_baseline
```

```python
def initialize_model():
    model = models.Sequential()
    
    model.add(layers.Dense(10, activation = 'relu', input_dim = X_train.shape[-1]))
    model.add(layers.Dense(7, activation = 'relu'))
    model.add(layers.Dense(2, activation = 'softmax'))
    
    return model
```

```python
model = initialize_model()
```

```python
model.summary()
```

```python
def compile_model(model, optimizer_name):
    model_compiled = model.compile(optimizer = optimizer_name,
                  loss = 'BinaryCrossentropy',
                  metrics='accuracy'
    )
    return model_compiled
```

```python
opt = Adam(
    learning_rate=0.01, beta_1=0.9, beta_2=0.99
)
```

```python
compile_model(model, opt)
```

```python
es = EarlyStopping(patience = 100, restore_best_weights=True)

history = model.fit(X_train, y_train, batch_size = 16, epochs = 500, validation_split = 0.2, callbacks = [es])
```

# Imports and preprocessing

## Imports

```python
import pandas as pd
```

## Teams DB

```python
data = pd.read_csv('data/england-premier-league-teams-2018-to-2019-stats.csv')
```

```python
data.head(5)
```

```python
start_data = data[['common_name','matches_played', 'wins', 'wins_home','wins_away','losses_home','losses_away', 'losses']]
```

```python
start_data['home_percentage'] = (start_data['wins_home']/19).round(2)
```

```python
start_data['away_percentage'] = (start_data['wins_away']/19).round(2)
```

```python
start_data['points'] = (data['points_per_game']*38).round(0)
```

```python
start_data.sort_values('points', ascending = False)
```

## Matches DB

```python
match_data = pd.read_csv('data/england-premier-league-matches-2018-to-2019-stats.csv')
```

```python
match_data.head(5)
```

```python
match_data_pre = match_data.drop('timestamp', axis = 1)
```

```python
match_data_pre = match_data.drop('home_ppg', axis = 1)
match_data_pre = match_data.drop('away_ppg', axis = 1)
```

```python
match_data_pre = match_data_pre[['home_team_name', 'away_team_name', 'home_team_goal_count', 'away_team_goal_count']]
```

```python
match_data_pre
```

```python
match_data_pre['away_team_away_percentage'] = 0
```

```python
for i in range(0, len(match_data_pre)):
    for j in range(0, len(start_data)):
        if match_data_pre['home_team_name'][i] == start_data['common_name'][j]:
            match_data_pre['home_team_home_percentage'][i] = start_data['home_percentage'][j]
            
        if match_data_pre['away_team_name'][i] == start_data['common_name'][j]:
            match_data_pre['away_team_away_percentage'][i] = start_data['away_percentage'][j]
```

```python
match_data_pre
```

```python
match_data_pre['home_win'] = 0
match_data_pre['away_win'] = 0
```

```python
for i in range(0,len(match_data_pre)):
    if match_data_pre['home_team_goal_count'][i] > match_data_pre['away_team_goal_count'][i]:
        match_data_pre['home_win'][i] = 1
        match_data_pre['away_win'][i] = 0
    elif match_data_pre['home_team_goal_count'][i] < match_data_pre['away_team_goal_count'][i]:
        match_data_pre['away_win'][i] = 1
        match_data_pre['home_win'][i] = 0

for i in range(0,len(match_data_pre)):
    if match_data_pre['home_win'][i] == 0 and match_data_pre['away_win'][i] == 0:
        match_data_pre['home_win'][i] = 0.5
        match_data_pre['away_win'][i] = 0.5
    
       
```
